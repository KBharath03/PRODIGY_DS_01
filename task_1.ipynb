Importing libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

Loading the dataset

```python
df= pd.read_csv('population.csv')
print(df)
```

Checking for null values.

```python
 df.isnull().sum()
```

Diaplaying the data types of each column

```python
df.dtypes
```

Printing the categorical columns in the dataset

```python
categorical_columns = df.select_dtypes(include=['object', 'category']).columns
print("Categorical columns:")
print(categorical_columns)
```

Displaying the continuous columns in the dataset

```python
continuous_columns = df.select_dtypes(include=['int64', 'float64']).columns
print("Continuous columns:")
print(continuous_columns)
```

Distribution of categorical columns(Bar graph)

```python
for column in categorical_columns:
    plt.figure(figsize=(100,6))
    sns.countplot(x=column, data=df, palette='viridis')
    plt.title(f'Distribution of {column}')
    plt.show()
```

Distribution of categorical columns(Histogram)

```python
for column in categorical_columns:
    plt.figure(figsize=(8, 6))
    plt.hist(df[column], bins=20, color='skyblue', edgecolor='black')
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
```

Distribution of continuous columns(Histogram)

```python
for column in continuous_columns:
    plt.figure(figsize=(8, 6))
    plt.hist(df[column], bins=20, color='skyblue', edgecolor='black')
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
```
